```python
df=pd.read_csv("student_data.csv")
```

```python
#Find the number of students who have completed both ssc and hsc with a GPA greater than 4

filtered=df[(df["ssc_gpa"]>4) & (df["hsc_gpa"]>4)]
print("The number of the students who have completed both:\n",filtered[["first_name","last_name","hsc_gpa","ssc_gpa"]].sample(10))
```

```python
#Q2 Display all students who have attended study abroad in Canada

display_abroad_canada=df[df["study_abroad_country"]=="Canada"]
print(display_abroad_canada[["first_name","last_name"]].sample(5))
```

```python
#Q5
#Create a pie chart showing the distribution of students by graduation_year
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data_file = "student_data.csv"
data = pd.read_csv(data_file)

# Ensure the 'Graduation Year' column exists
if 'graduation_date' in data.columns:
    # Count the occurrences of each graduation year
    year_counts = data['graduation_date'].value_counts()

    # Plot the pie chart
    plt.figure(figsize=(10, 10))
    plt.pie(year_counts, labels=year_counts.index, startangle=140, colors=plt.cm.tab20.colors)
    plt.title("Distribution of Students by Graduation Year")
    plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.

    # Show the chart
    plt.show()
else:
    print("The dataset does not contain a 'Graduation Year' column.")
```
